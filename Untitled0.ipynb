```python
# Открываем PDF файл и извлекаем текст из него

text = textract.process('/content/drive/MyDrive/Colab Notebooks/Парсинг/he-qi-zheng-mof-808-a-metal-organic-framework-with.pdf').decode('utf-8')

# Задаем таргетные слова
target_words = ['pH']

# Ищем таргетные слова в тексте
found_words = [word for word in target_words if re.search(word, text)]

# Разбиваем текст на строки и удаляем пустые строки
lines = [line.strip() for line in text.split('\n') if line.strip()]

# Удаляем строки, в которых нет таргетных слов
lines = [line for line in lines if any([re.search(word, line.lower()) for word in found_words])]

# Сжимаем текст
compressed_text = '\n'.join(lines)

print(compressed_text)
```

```python
# Пользовательские таргетные слова
target_words = ['pH']

# Регулярное выражение для поиска предложений с таргетными словами
pattern = re.compile(r'([A-Z][^\.!?]*?(?:' + '|'.join(target_words) + ')[^\.!?]*?[\.!?])', re.M)

# Поиск предложений с таргетными словами в тексте
found_sentences = re.findall(pattern, text)

# Вывод найденных предложений
for sentence in found_sentences:
    print(sentence, end='\n')
```

```python
# Открываем PDF-файл и создаем объект для чтения
!pip install PyPDF2
import PyPDF2
pdf_file = open('/content/drive/MyDrive/Colab Notebooks/Парсинг/he-qi-zheng-mof-808-a-metal-organic-framework-with.pdf', 'rb')
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Получаем текст из всех страниц PDF-файла
text = ''
for page_num in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[page_num]
    text += page.extract_text()

# Пользовательские таргетные слова
target_words = ['pH']

# Регулярное выражение для поиска слов в тексте
pattern = re.compile(r'([^.]*?\b(' + '|'.join(target_words) + r')\b[^.]*\.)', re.IGNORECASE | re.DOTALL)

# Поиск предложений с таргетными словами в тексте PDF-файла
found_sentences = re.findall(pattern, text)

# Вывод найденных предложений
for sentence in found_sentences:
    print(sentence[0], sep='\n', end='\n')
```

```python

def find_caps_lock_headings(filename):
  headings = []

  pdf_file = open('/content/drive/MyDrive/Colab Notebooks/Парсинг/he-qi-zheng-mof-808-a-metal-organic-framework-with.pdf', 'rb')
  pdf_reader = PyPDF2.PdfReader(pdf_file)

  for page_num in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[page_num]
    text = page.extract_text()

# Используем регулярное выражение для поиска заголовков
  pattern = r'\b[A-Z\s]+\b'
  matches = re.findall(pattern, text)

# Добавляем найденные заголовки в список
  for match in matches:
    if match not in headings:
      headings.append(match)

  return headings

# Пример использования функции
filename = '/content/drive/MyDrive/Colab Notebooks/Парсинг/he-qi-zheng-mof-808-a-metal-organic-framework-with.pdf'
headings = find_caps_lock_headings(filename)

if headings:
  print('Найденные заголовки:')
for heading in headings:
  print(heading)
else:
  print('Заголовки не найдены.')
```
